```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import random
```

```python
x=[random.randint(0,1000) for i in range(1000)]
y=[random.randint(0,1000) for i in range(1000)]

x=np.asanyarray(x)
y=np.asanyarray(y)
```

```python
df=pd.DataFrame({'x1':x,'x2':y,'y':x+y})
df
```

```python
xtr,xte,ytr,yte=train_test_split(df[['x1','x2']],df['y'],test_size=0.2)
```

```python
xtr.shape
```

```python
ytr.shape
```

```python
model =keras.Sequential()
model.add(keras.layers.Dense(6,activation='relu',input_shape=(2,)))
model.add(keras.layers.Dense(6,activation='relu'))

model.add(keras.layers.Dense(1,activation='relu'))

model.compile(optimizer='adam',loss='mse',metrics=['mae'])
model.fit(xtr,ytr,epochs=350)
```

```python
ypred=model.predict(xte)
```

```python
ypred.round()
```

```python
from sklearn.metrics import mean_squared_error
1-mean_squared_error(yte,ypred)
```
